```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Define an input tensor with some integer values
input_tensor = torch.tensor([[1, 3, 2, 4],
                             [5, 6, 1, 2],
                             [7, 8, 3, 1],
                             [1, 2, 0, 3]], dtype=torch.float32)

print(input_tensor.shape)  # size of the input
# add two additional dimensions to create a batch for nn methods to be used
batch_input_tensor = input_tensor.unsqueeze(0).unsqueeze(0)
print(batch_input_tensor.shape)
# Define the max pooling layer, stride of the window.
max_pool = nn.MaxPool2d(kernel_size=2, stride=2)

# Applying max pooling
output_tensor = max_pool(batch_input_tensor)

# Plot the input and output tensors
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the input
''' squeeze returns a tensor with all specified dimensions
of input of size 1 removed from the left hand side.'''
axs[0].imshow(batch_input_tensor.squeeze(), cmap='brg')
axs[0].set_title('Batch Input Tensor')
axs[0].axis('off')

# Annotate values on the input plot
for i in range(batch_input_tensor.shape[2]):# dimension x or 1st
    for j in range(batch_input_tensor.shape[3]): # dimension y or 2nd
        # first two values [,,,] are 0 and 0
        axs[0].text(j, i, int(batch_input_tensor[0, 0, i, j].item()), ha='center', va='center', color='white')

# output tensor plotting
axs[1].imshow(output_tensor.squeeze(), cmap='brg')
axs[1].set_title('Max-Pooled Feature Map')
axs[1].axis('off')

# Annotate output tensor values in the plot
for i in range(output_tensor.shape[2]):
    for j in range(output_tensor.shape[3]):
        axs[1].text(j, i, int(output_tensor[0, 0, i, j].item()), ha='center', va='center', color='white')

plt.show()

```
